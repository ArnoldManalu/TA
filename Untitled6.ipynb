```python
# Install dan Import Libraries
!pip install timm
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
import timm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import torch.nn.functional as F  # Untuk LabelSmoothingLoss
from torch.optim.lr_scheduler import ReduceLROnPlateau
import warnings
import random  # Untuk reproducibility
warnings.filterwarnings('ignore')

# Set seed untuk semua library agar hasil konsisten
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

# Set deterministic mode untuk CuDNN (membuat operasi GPU deterministic)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# Set environment variable untuk hashing reproducibility
os.environ['PYTHONHASHSEED'] = str(SEED)

print(f"‚úÖ Seeds diset ke {SEED} untuk reproducibility penuh.")

# Definisi LabelSmoothingLoss (custom implementation)
class LabelSmoothingLoss(nn.Module):
    def __init__(self, num_classes, smoothing=0.0):
        super(LabelSmoothingLoss, self).__init__()
        self.num_classes = num_classes
        self.smoothing = smoothing

    def forward(self, input, target):
        log_prob = F.log_softmax(input, dim=-1)
        weight = input.new_ones(input.size()) * self.smoothing / (self.num_classes - 1.)
        weight.scatter_(1, target.unsqueeze(1), (1. - self.smoothing) / float(self.num_classes))
        loss = (-weight * log_prob).sum(dim=1).mean()
        return loss

# (Colab drive mount removed)
```

```python
# 1. Define dataset path and classes
# Use a local dataset path (update as needed). Example: './data/klasifikasi_katarak'
dataset_path = r'D:/PERKULIAHAN/SEMESTER 5/klasifikasi_katarak'
classes = ['Immature', 'Mature', 'Normal']
num_classes = len(classes)  # 3 classes

# Verify dataset structure
print("# Verifikasi Dataset")
total_images = 0
for class_name in classes:
    class_path = os.path.join(dataset_path, class_name)
    if os.path.exists(class_path):
        num_images = len([f for f in os.listdir(class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
        print(f"{class_name}: {num_images} images")
        total_images += num_images
    else:
        print(f"Peringatan: {class_path} tidak ditemukan")
print(f"Total images: {total_images}")
```

```python
# 2. Define transforms
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
common_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=mean, std=std)
])
```

```python
# 3. Load full dataset dengan common transform
full_dataset = ImageFolder(dataset_path, transform=common_transform)

# Split dataset menjadi 70% train, 20% val, 10% test (stratified)
train_idx, temp_idx = train_test_split(
    np.arange(len(full_dataset)),
    test_size=0.3,
    stratify=full_dataset.targets,
    random_state=SEED  # Fixed seed
)
val_idx, test_idx = train_test_split(
    temp_idx,
    test_size=1/3,  # dari 30% sisa ‚Üí 10% test dan 20% val
    stratify=[full_dataset.targets[i] for i in temp_idx],
    random_state=SEED  # Fixed seed
)

# Buat subset untuk masing-masing bagian
train_dataset = Subset(full_dataset, train_idx)
val_dataset = Subset(full_dataset, val_idx)
test_dataset = Subset(full_dataset, test_idx)

# Tampilkan ukuran tiap subset
print(f"# Dataset size: Train = {len(train_dataset)}, Val = {len(val_dataset)}, Test = {len(test_dataset)}")
```

```python
# 4. Define hyperparameters
batch_size = 16
epochs = 100
learning_rate = 0.001
dropout_rate = 0.5
optimizer_name = 'AdamW'
units = 64
print(f"\n# Current Hyperparameters:")
print(f"Batch Size: {batch_size}, Epochs: {epochs}, Learning Rate: {learning_rate}")
print(f"Dropout Rate: {dropout_rate}, Optimizer: {optimizer_name}, Units: {units}")

def seed_worker(worker_id):
    worker_seed = torch.initial_seed() % 2**32
    np.random.seed(worker_seed)
    random.seed(worker_seed)

g_seed_worker = seed_worker
```

```python
# 5. Create data loaders dengan seed_worker untuk shuffle reproducibility
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, worker_init_fn=g_seed_worker, generator=torch.Generator().manual_seed(SEED))
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
```

```python
# 6. Define model creation function
def create_model(
    model_name='mobilenetv4_hybrid_medium.e500_r224_in1k',
    num_classes=3,
    dropout_rate=0.3,
    units=256
):
    # Buat backbone pretrained tanpa classifier
    model = timm.create_model(
        model_name,
        pretrained=True,
        num_classes=0,  # agar bisa menambahkan head sendiri
        drop_rate=dropout_rate * 0.5  # sedikit regulasi di backbone
    )
    # Hitung jumlah fitur output backbone secara dinamis
    model.eval()  # pastikan tidak ada efek dropout/batchnorm
    with torch.no_grad():
        dummy_input = torch.randn(1, 3, 224, 224)
        features = model.forward_features(dummy_input)
        pre_logits = model.forward_head(features, pre_logits=True)
        num_features = pre_logits.shape[1]
    print(f"# Model {model_name}: Detected num_features = {num_features}")
    # Definisikan classifier baru (inisialisasi linear layer deterministic karena seed global)
    classifier = nn.Sequential(
        nn.Dropout(dropout_rate),
        nn.Linear(num_features, units),
        nn.ReLU(inplace=True),
        nn.Dropout(dropout_rate),
        nn.Linear(units, num_classes)
    )
    # Ganti classifier bawaan model dengan classifier baru
    model.classifier = classifier
    # Kembalikan model yang siap digunakan
    return model
```

```python
# 7. Training function dengan Early Stopping dan Gradient Clipping
def train_model(model, train_loader, val_loader, test_loader, epochs, learning_rate, optimizer_name, device, num_classes, classes, save_path='./percobaan2.pth'):
    print("üöÄ Training dimulai...\n")
    # Gunakan Label Smoothing untuk regularisasi ringan
    criterion = LabelSmoothingLoss(num_classes, smoothing=0.1)
    # Optimizer
    if optimizer_name == 'AdamW':
        optimizer = optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=1e-4)
    elif optimizer_name == 'SGD':
        optimizer = optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9, weight_decay=1e-4)
    else:
        raise ValueError(f"Optimizer {optimizer_name} tidak dikenali.")
    # Scheduler: turunkan LR jika val_loss stagnan
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)
    # Early stopping setup
    patience = 7
    min_delta = 0.001
    best_val_loss = float('inf')
    early_stop_counter = 0
    best_model_state = None
    # Logging list
    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    for epoch in range(epochs):
        model.train()
        running_loss, correct, total = 0.0, 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            # Gradient clipping ‚Üí mencegah exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
        # Hitung metrik training
        train_loss = running_loss / len(train_loader)
        train_acc = 100. * correct / total
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        # --- Validation Phase ---
        model.eval()
        val_loss, v_correct, v_total = 0.0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = outputs.max(1)
                v_total += labels.size(0)
                v_correct += predicted.eq(labels).sum().item()
        val_loss /= len(val_loader)
        val_acc = 100. * v_correct / v_total
        val_losses.append(val_loss)
        val_accs.append(val_acc)
        scheduler.step(val_loss)
        print(f"Epoch [{epoch+1}/{epochs}] | "
              f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% || "
              f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
        # --- Early Stopping ---
        if val_loss < best_val_loss - min_delta:
            best_val_loss = val_loss
            early_stop_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            early_stop_counter += 1
            if early_stop_counter >= patience:
                print(f"‚èπÔ∏è Early stopping di epoch {epoch+1}")
                break
    # --- Evaluasi di Test Set ---
    model.eval()
    test_correct, test_total = 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = outputs.max(1)
            test_total += labels.size(0)
            test_correct += predicted.eq(labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    test_acc = 100. * test_correct / test_total
    report = classification_report(all_labels, all_preds, target_names=classes, output_dict=True)
    # --- Simpan Model Terbaik ---
    if best_model_state:
        model.load_state_dict(best_model_state)
        torch.save(best_model_state, save_path)
        print(f"\nüíæ Model terbaik disimpan ke: {save_path}")
    print(f"\n‚úÖ Training selesai. Test Accuracy: {test_acc:.2f}%")
    return {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'train_accs': train_accs,
        'val_accs': val_accs,
        'test_acc': test_acc,
        'f1_macro': report['macro avg']['f1-score'],
        'classification_report': report,
        'all_preds': all_preds,
        'all_labels': all_labels,
        'best_val_acc': max(val_accs) if val_accs else 0
    }
```

```python
# 8. Set device, create/train model, dan evaluasi
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\n# Menggunakan device: {device}")

# Buat model (deterministic karena seed global)
model = create_model(num_classes=num_classes, dropout_rate=dropout_rate, units=units).to(device)
print("\n# Memulai Training...")
metrics = train_model(
    model, train_loader, val_loader, test_loader,
    epochs, learning_rate, optimizer_name,
    device, num_classes, classes
)

# Ringkasan Hasil Akhir Setelah Training
print("\n" + "="*60)
print("üìä RINGKASAN HASIL AKHIR TRAINING")
print("="*60)
print(f"Final Train Loss:     {metrics['train_losses'][-1]:.4f}")
print(f"Final Val Loss:       {metrics['val_losses'][-1]:.4f}")
print(f"Final Train Acc:      {metrics['train_accs'][-1]:.2f}%")
print(f"Final Val Acc:        {metrics['val_accs'][-1]:.2f}%")
print(f"Best Val Acc (selama training): {metrics['best_val_acc']:.2f}%")
print(f"Test Accuracy Akhir:  {metrics['test_acc']:.2f}%")
print(f"F1-Macro Score:       {metrics['f1_macro']:.4f}")
print("="*60 + "\n")

# Ambil hasil prediksi dan label
all_preds = metrics['all_preds']
all_labels = metrics['all_labels']

# Evaluasi performa (classification report)
print("\n# Laporan Klasifikasi (Test Set):")
print(classification_report(all_labels, all_preds, target_names=classes, zero_division=0))  # untuk hindari error jika ada kelas tanpa prediksi

# Plot Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Test Set)')
plt.tight_layout()
plt.show()
```

```python
# 9. Testing Model pada Data Test
print("\n# Evaluasi Tambahan pada Data Test (menggunakan model terbaik)...")
# Model sudah loaded dengan best state di train_model, jadi langsung evaluasi ulang untuk konfirmasi
model.eval()
all_preds_test = []
all_labels_test = []
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        all_preds_test.extend(predicted.cpu().numpy())
        all_labels_test.extend(labels.cpu().numpy())

# Hitung metrik evaluasi (harus sama dengan di training)
test_acc_confirm = accuracy_score(all_labels_test, all_preds_test) * 100
test_report_confirm = classification_report(all_labels_test, all_preds_test, target_names=classes, zero_division=0)
test_cm_confirm = confusion_matrix(all_labels_test, all_preds_test)

# Tampilkan hasil konfirmasi
print(f"‚úÖ Konfirmasi Akurasi Test: {test_acc_confirm:.2f}% (sama dengan hasil training)")
print("\n# Laporan Klasifikasi Konfirmasi:")
print(test_report_confirm)

# Plot Confusion Matrix Konfirmasi
plt.figure(figsize=(8, 6))
sns.heatmap(test_cm_confirm, annot=True, fmt='d', cmap='Greens',
            xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Konfirmasi Test Set')
plt.tight_layout()
plt.show()
```

```python
# 10. Prediksi Satu Gambar (Single Image Inference)
from PIL import Image

def predict_single_image(model, image_path, classes, device, transform):
    if not os.path.exists(image_path):
        print(f"‚ùå File gambar '{image_path}' tidak ditemukan. Lewati prediksi single image.")
        return
    # Buka gambar dan konversi ke RGB
    image = Image.open(image_path).convert('RGB')
    # Terapkan transformasi yang sama seperti training
    img_tensor = transform(image).unsqueeze(0).to(device)  # tambah dimensi batch
    # Prediksi
    model.eval()
    with torch.no_grad():
        output = model(img_tensor)
        probs = torch.softmax(output, dim=1)[0]
        pred_class = torch.argmax(probs).item()
    # Tampilkan hasil
    print(f"\nüñºÔ∏è Gambar: {os.path.basename(image_path)}")
    for i, cls in enumerate(classes):
        print(f"{cls:10s}: {probs[i]*100:.2f}%")
    print(f"\n‚úÖ Prediksi Akhir: {classes[pred_class]} ({probs[pred_class]*100:.2f}% confidence)")
    # Tampilkan gambar dengan prediksi
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.title(f"Prediksi: {classes[pred_class]} ({probs[pred_class]*100:.1f}%)")
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Update image_path to a local image (e.g., './arn.jpg')
image_path = './arn.jpg'
predict_single_image(model, image_path, classes, device, common_transform)

print("\nüéâ Training dan evaluasi selesai. Ringkasan akhir telah ditampilkan di atas!")
```
