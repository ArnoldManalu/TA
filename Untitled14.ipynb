```python
import os
import shutil
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torch.amp import autocast, GradScaler
from torchvision import transforms
from torchvision.datasets import ImageFolder
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
from roboflow import Roboflow
import timm
from tqdm import tqdm
import optuna
from optuna.storages import RDBStorage
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
# 1. FOLDER HASIL
TIMESTAMP = datetime.now().strftime("%Y%m%d_%H%M%S")
RESULT_DIR = Path("/content/cataract_results") / TIMESTAMP
RESULT_DIR.mkdir(parents=True, exist_ok=True)
(RESULT_DIR / "models").mkdir(exist_ok=True)
(RESULT_DIR / "plots").mkdir(exist_ok=True)
CSV_PATH = RESULT_DIR / "all_trials_summary.csv"
if not CSV_PATH.exists():
    pd.DataFrame(columns=[
        "trial_number","batch_size","lr","weight_decay","dropout_rate","fc_units",
        "best_epoch","best_val_acc","best_val_loss","best_train_acc","best_train_loss",
        "final_test_acc","train_size","val_size","test_size","total_epochs","early_stopped","timestamp"
    ]).to_csv(CSV_PATH, index=False)
print(f"Hasil disimpan di: {RESULT_DIR}\n")
```

```python
# 2. SPLIT 80/10/10 PER KELAS
split_root = Path("/content/cataract_dataset_split")
if split_root.exists() and len(list((split_root / "train").glob("*/*"))) > 0:
    print(f"Dataset sudah ada → menggunakan existing split\n")
else:
    print("Membuat split 80/10/10 PER KELAS dari Roboflow...")
    if split_root.exists():
        shutil.rmtree(split_root)
    rf = Roboflow(api_key="QOd5ldAdjiaehHn5m6WC")
    project = rf.workspace("dentalogic8").project("katarak-b97j5")
    version = project.version(1)
    dataset = version.download("folder")
    original_train_path = Path(dataset.location) / "train"
    split_root.mkdir(parents=True)
    for s in ["train", "val", "test"]:
        (split_root / s).mkdir(exist_ok=True)
    classes = ['Normal', 'Immature', 'Mature']
    total_train = total_val = total_test = 0
    for cls in classes:
        class_path = original_train_path / cls
        if not class_path.exists() or len(list(class_path.glob("*.*"))) == 0:
            print(f"Kelas {cls} tidak ada!")
            continue
        files = list(class_path.glob("*.*"))
        n = len(files)
        print(f"{cls}: {n} gambar")
        train_files, temp = train_test_split(files, test_size=0.2, random_state=42, shuffle=True)
        val_files, test_files = train_test_split(temp, test_size=0.5, random_state=42, shuffle=True)
        for s, flist in zip(["train", "val", "test"], [train_files, val_files, test_files]):
            dest_dir = split_root / s / cls
            dest_dir.mkdir(parents=True, exist_ok=True)
            for f in flist:
                shutil.copy(f, dest_dir / f.name)
        total_train += len(train_files)
        total_val += len(val_files)
        total_test += len(test_files)
    print(f"\nSELESAI! Split 80/10/10 per kelas:")
    print(f" Train : {total_train} gambar")
    print(f" Val : {total_val} gambar")
    print(f" Test : {total_test} gambar\n")
```

```python
# 3. LOAD DATASET
mean, std = [0.485, 0.456, 0.406], [0.229, 0.224, 0.225]
data_transforms = {
    'train': transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean, std)]),
    'val':   transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean, std)]),
    'test':  transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean, std)])
}

train_ds = ImageFolder(split_root / "train", data_transforms['train'])
val_ds   = ImageFolder(split_root / "val",   data_transforms['val'])
test_ds  = ImageFolder(split_root / "test",  data_transforms['test'])

print(f"JUMLAH DATASET FINAL:")
print(f"   Train : {len(train_ds):4d} gambar")
print(f"   Val   : {len(val_ds):4d} gambar")
print(f"   Test  : {len(test_ds):4d} gambar\n")
```

```python
# 4. SEED & DEVICE
def set_seed(seed=42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.benchmark = True
    os.environ['PYTHONHASHSEED'] = str(seed)
set_seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}\n")
```

```python
# 5. MODEL
class MobileNetV4Custom(nn.Module):
    def __init__(self, dropout_rate=0.3, fc_units=256):
        super().__init__()
        self.backbone = timm.create_model("mobilenetv4_conv_blur_medium.e500_r224_in1k", pretrained=True, num_classes=0)
        self.head = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(1280, fc_units),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate * 0.5),
            nn.Linear(fc_units, 3)
        )
    def forward(self, x):
        return self.head(self.backbone(x))
```

```python
# 6. TRAINING
def train_model(trial):
    bs = trial.suggest_categorical("batch_size", [32, 64, 128])
    lr = trial.suggest_float("lr", 1e-5, 5e-3, log=True)
    wd = trial.suggest_float("weight_decay", 1e-6, 1e-2, log=True)
    do = trial.suggest_float("dropout_rate", 0.2, 0.5, step=0.05)
    fc = trial.suggest_categorical("fc_units", [256, 512])
    set_seed(42)
    train_loader = DataLoader(train_ds, batch_size=bs, shuffle=True,
                              num_workers=16, pin_memory=True, persistent_workers=True, prefetch_factor=2)
    val_loader = DataLoader(val_ds, batch_size=bs*2, shuffle=False,
                              num_workers=16, pin_memory=True, persistent_workers=False)
    test_loader = DataLoader(test_ds, batch_size=128, shuffle=False,
                              num_workers=16, pin_memory=True, persistent_workers=False)
    model = MobileNetV4Custom(dropout_rate=do, fc_units=fc).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=wd)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=100)
    scaler = GradScaler()
    history = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}
    best_val_acc = 0.0
    best_val_loss = float('inf')
    best_train_acc = 0.0
    best_train_loss = float('inf')
    best_epoch = 0
    patience = 7
    no_improve = 0
    best_val_preds = best_val_labels = None
    best_test_preds = best_test_labels = None
    trial_dir = RESULT_DIR / "plots" / f"trial_{trial.number:03d}"
    trial_dir.mkdir(parents=True, exist_ok=True)
    print(f"TRIAL {trial.number:03d} | BS:{bs} LR:{lr:.2e} WD:{wd:.2e} DO:{do:.2f} FC:{fc}")
    print("Ep Train Loss Train Acc Val Loss Val Acc Status")
    print("━" * 64)
    for epoch in range(100):
        # ================== TRAIN ==================
        model.train()
        train_loss_total = 0.0
        train_correct = 0
        train_total = 0
        for x, y in tqdm(train_loader, desc=f"Ep {epoch+1:3d} [TRAIN]", leave=False, colour="green"):
            x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
            optimizer.zero_grad(set_to_none=True)
            with autocast(device_type='cuda'):
                out = model(x)
                loss = criterion(out, y)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            # Handle potential NaN in train loss (clip or log)
            if torch.isnan(loss):
                print(f"Warning: NaN detected in train loss at epoch {epoch+1}. Skipping update.")
                continue
            train_loss_total += loss.item() * y.size(0)
            pred = out.argmax(1)
            train_correct += (pred == y).sum().item()
            train_total += y.size(0)
        if train_total == 0:
            train_acc = 0.0
            train_loss = float('inf')
        else:
            train_acc = train_correct / train_total
            train_loss = train_loss_total / train_total
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        # Update best train metrics
        if train_acc > best_train_acc:
            best_train_acc = train_acc
            best_train_loss = train_loss
        # ================== VALIDATION ==================
        model.eval()
        val_loss_total = 0.0
        val_correct = 0
        val_total = 0
        val_preds, val_labels = [], []
        with torch.no_grad():
            # Disable autocast for validation to prevent fp16 overflow/NaN issues
            for x, y in tqdm(val_loader, desc=f"Ep {epoch+1:3d} [VAL ]", leave=False, colour="blue"):
                x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
                # Compute in full precision to avoid NaN
                out = model(x)
                loss = criterion(out, y)
                # Handle NaN explicitly
                loss_item = loss.item()
                if np.isnan(loss_item):
                    print(f"Warning: NaN detected in val loss at epoch {epoch+1}, batch. Using previous loss value.")
                    loss_item = history['val_loss'][-1] if history['val_loss'] else float('inf')
                val_loss_total += loss_item * y.size(0)
                pred = out.argmax(1)
                val_correct += (pred == y).sum().item()
                val_total += y.size(0)
                val_preds.extend(pred.cpu().numpy())
                val_labels.extend(y.cpu().numpy())
        if val_total == 0:
            val_acc = 0.0
            val_loss = float('inf')
        else:
            val_acc = val_correct / val_total
            val_loss = val_loss_total / val_total
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        scheduler.step()
        # Update best val metrics
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_val_loss = val_loss
            best_epoch = epoch + 1
            # PERBAIKAN: Simpan SELURUH MODEL (bukan hanya state_dict)
            torch.save(model, RESULT_DIR / "models" / f"trial_{trial.number:03d}_best.pth")
            no_improve = 0
            best_val_preds = val_preds
            best_val_labels = val_labels
            # Test saat BEST (also without autocast)
            model.eval()
            best_test_preds, best_test_labels = [], []
            with torch.no_grad():
                for x, y in test_loader:
                    x_dev = x.to(device, non_blocking=True)
                    out_test = model(x_dev)
                    pred_test = out_test.argmax(1)
                    best_test_preds.extend(pred_test.cpu().numpy())
                    best_test_labels.extend(y.numpy()) # y is already CPU tensor
        else:
            no_improve += 1
        # Print with safe formatting to avoid NaN display issues
        print(f"{epoch+1:3d} {train_loss:8.4f} {train_acc:8.4f} {val_loss:8.4f} {val_acc:8.4f} {'BEST' if val_acc > best_val_acc else ''}")
        if no_improve >= patience:
            print(f"EARLY STOPPING! Best Val Acc: {best_val_acc:.5f} @ epoch {best_epoch}\n")
            break
    # ================== PLOT & CONFUSION MATRIX ==================
    trial_dir = RESULT_DIR / "plots" / f"trial_{trial.number:03d}"
    trial_dir.mkdir(parents=True, exist_ok=True)
    # Accuracy
    plt.figure(figsize=(8, 6))
    plt.plot(history['train_acc'], label='Train Acc', color='blue')
    plt.plot(history['val_acc'], label='Val Acc', color='orange')
    plt.title(f"Trial {trial.number} - Accuracy")
    plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend(); plt.grid(alpha=0.3)
    plt.savefig(trial_dir / "accuracy_curve.png", dpi=200, bbox_inches='tight')
    plt.close()
    # Loss (handle NaN in plotting by replacing with inf or last value)
    safe_train_loss = [x if not np.isnan(x) else float('inf') for x in history['train_loss']]
    safe_val_loss = [x if not np.isnan(x) else float('inf') for x in history['val_loss']]
    plt.figure(figsize=(8, 6))
    plt.plot(safe_train_loss, label='Train Loss', color='blue')
    plt.plot(safe_val_loss, label='Val Loss', color='orange')
    plt.title("Loss Curve")
    plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend(); plt.grid(alpha=0.3)
    plt.savefig(trial_dir / "loss_curve.png", dpi=200, bbox_inches='tight')
    plt.close()
    # Validation Confusion Matrix
    if best_val_preds:
        cm_val = confusion_matrix(best_val_labels, best_val_preds)
        plt.figure(figsize=(7, 6))
        sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues',
                    xticklabels=['Normal','Immature','Mature'], yticklabels=['Normal','Immature','Mature'])
        plt.title(f"Validation Confusion Matrix\nAcc: {best_val_acc:.5f}")
        plt.savefig(trial_dir / "val_confusion_matrix.png", dpi=200, bbox_inches='tight')
        plt.close()
    # Test Confusion Matrix
    if best_test_preds:
        cm_test = confusion_matrix(best_test_labels, best_test_preds)
        test_acc = np.trace(cm_test) / np.sum(cm_test)
        plt.figure(figsize=(7, 6))
        sns.heatmap(cm_test, annot=True, fmt='d', cmap='Greens',
                    xticklabels=['Normal','Immature','Mature'], yticklabels=['Normal','Immature','Mature'])
        plt.title(f"Test Confusion Matrix\nAcc: {test_acc:.5f}")
        plt.savefig(trial_dir / "test_confusion_matrix.png", dpi=200, bbox_inches='tight')
        plt.close()
    else:
        test_acc = 0.0
    # ================== CSV 100% AMAN ==================
    row = {
        "trial_number": trial.number, "batch_size": int(bs), "lr": round(lr, 10),
        "weight_decay": round(wd, 10), "dropout_rate": float(do), "fc_units": int(fc),
        "best_epoch": best_epoch, "best_val_acc": round(best_val_acc, 5),
        "best_val_loss": round(best_val_loss, 5), "best_train_acc": round(best_train_acc, 5),
        "best_train_loss": round(best_train_loss, 5), "final_test_acc": round(test_acc, 5),
        "train_size": len(train_ds), "val_size": len(val_ds), "test_size": len(test_ds),
        "total_epochs": len(history['train_acc']),
        "early_stopped": "True" if no_improve >= patience else "False",
        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    pd.DataFrame([row]).to_csv(CSV_PATH, mode='a', header=False, index=False)
    print(f"SELESAI! Semua disimpan di: {trial_dir}\n")
    return best_val_acc
```

```python
# 7. JALANKAN OPTUNA
storage = RDBStorage(url=f"sqlite:///{RESULT_DIR}/optuna_study.db")
study = optuna.create_study(direction="maximize", sampler=optuna.samplers.TPESampler(seed=42),
                            pruner=optuna.pruners.MedianPruner(), storage=storage, load_if_exists=True)
print("MULAI OPTUNA – 100% AMAN, TIDAK ADA NaN, LOKAL RUNTIME!")
study.optimize(train_model, n_trials=50)
```
